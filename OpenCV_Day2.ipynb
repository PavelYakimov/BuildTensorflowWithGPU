```python
shadowedImage = cv2.imread('shadow.jpg')
plt.figure(figsize=(15,15))
plt.imshow(shadowedImage)
plt.show()
```

```python

ret, binary = cv2.threshold(cv2.cvtColor(shadowedImage, cv2.COLOR_BGR2GRAY), 150, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(15,15))
plt.imshow(binary, cmap='gray')
plt.show()
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

grayShadowImage = cv2.cvtColor(shadowedImage, cv2.COLOR_BGR2GRAY)

img = cv.medianBlur(grayShadowImage,5)

th2 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY,11,2)
th3 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY,11,2)

plt.figure(figsize=(15,15))
plt.imshow(grayShadowImage, cmap='gray')
plt.show()

plt.figure(figsize=(15,15))
plt.imshow(img, cmap='gray')
plt.show()

plt.figure(figsize=(15,15))
plt.imshow(binary, cmap='gray')
plt.show()

plt.figure(figsize=(15,15))
plt.imshow(th2, cmap='gray')
plt.show()

plt.figure(figsize=(15,15))
plt.imshow(th3, cmap='gray')
plt.show()
```
